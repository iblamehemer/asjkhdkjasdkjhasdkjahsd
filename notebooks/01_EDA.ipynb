```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from wordcloud import WordCloud
import os, warnings
warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize'] = (12, 5)
sns.set_theme(style='darkgrid')
print('Imports OK')
```

## 1. Slogan Dataset EDA

```python
np.random.seed(42)
industries = ['Tech','Fashion','Food','Finance','Health','Education','Travel','Sports']
tones = ['Bold','Formal','Youthful','Inspirational','Luxury','Friendly']
slogans_list = ['Just Do It','Think Different','Open Happiness','The Best or Nothing',
                'Finger Lickin Good','Impossible Is Nothing','Eat Fresh','Dream Big',
                'Power Your Potential','Make It Happen','Life is Good','Beyond Limits'] * 42
slogan_df = pd.DataFrame({
    'slogan': slogans_list[:500],
    'industry': np.random.choice(industries, 500),
    'tone': np.random.choice(tones, 500),
    'target_audience': np.random.choice(['B2B','B2C','Youth','Professionals'], 500),
    'word_count': np.random.randint(2, 8, 500),
    'rating': np.random.uniform(3.0, 5.0, 500).round(1)
})
print(f'Slogan Dataset: {slogan_df.shape}')
print(slogan_df.isnull().sum())
slogan_df.head()
```

```python
# VIZ 1: Industry & Tone Distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
industry_counts = slogan_df['industry'].value_counts()
axes[0].bar(industry_counts.index, industry_counts.values, color=plt.cm.Set2(np.linspace(0,1,len(industry_counts))))
axes[0].set_title('Slogan Count by Industry', fontsize=14, fontweight='bold')
axes[0].tick_params(axis='x', rotation=45)
tone_counts = slogan_df['tone'].value_counts()
axes[1].pie(tone_counts.values, labels=tone_counts.index, autopct='%1.1f%%')
axes[1].set_title('Tone Distribution', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('eda_viz1_slogan_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# VIZ 2: Word Cloud
text = ' '.join(slogan_df['slogan'].astype(str))
wc = WordCloud(width=1200, height=400, background_color='#1a1a2e', colormap='plasma', max_words=80).generate(text)
plt.figure(figsize=(14, 5))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title('Most Frequent Brand Slogan Words', fontsize=16, fontweight='bold')
plt.savefig('eda_viz2_wordcloud.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 2. Marketing Campaign Dataset EDA

```python
platforms = ['Instagram','Facebook','Twitter/X','LinkedIn','TikTok']
objectives = ['Brand Awareness','Engagement','Conversion','Lead Generation']
regions = ['North America','Europe','Asia','Middle East','Africa','South America']
campaign_df = pd.DataFrame({
    'platform': np.random.choice(platforms, 1000),
    'objective': np.random.choice(objectives, 1000),
    'region': np.random.choice(regions, 1000),
    'industry': np.random.choice(['Tech','Fashion','Food','Finance','Health'], 1000),
    'budget_usd': np.random.randint(500, 50000, 1000),
    'duration_days': np.random.randint(7, 90, 1000),
    'ctr_percent': np.random.uniform(0.5, 8.0, 1000).round(2),
    'roi_percent': np.random.uniform(-10, 300, 1000).round(1),
    'engagement_score': np.random.uniform(10, 100, 1000).round(1),
    'impressions': np.random.randint(1000, 500000, 1000),
    'conversions': np.random.randint(5, 5000, 1000)
})
print(f'Campaign Dataset: {campaign_df.shape}')
campaign_df.describe()
```

```python
# VIZ 3: KPI Distributions
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
for ax, col, color, label in zip(axes,
    ['ctr_percent','roi_percent','engagement_score'],
    ['#2E86AB','#F18F01','#44BBA4'],
    ['CTR (%)','ROI (%)','Engagement Score']):
    ax.hist(campaign_df[col], bins=30, color=color, edgecolor='white', alpha=0.85)
    ax.axvline(campaign_df[col].mean(), color='red', linestyle='--', label=f'Mean: {campaign_df[col].mean():.1f}')
    ax.set_title(f'{label} Distribution', fontweight='bold')
    ax.legend()
plt.suptitle('Campaign KPI Distributions', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig('eda_viz3_kpi_distributions.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# VIZ 4: Correlation Heatmap
numeric_cols = ['budget_usd','duration_days','ctr_percent','roi_percent','engagement_score','impressions','conversions']
corr = campaign_df[numeric_cols].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0, square=True, linewidths=0.5)
plt.title('Feature Correlation Matrix — Campaign Dataset', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('eda_viz4_correlation.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# VIZ 5: Platform vs CTR Boxplot
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
campaign_df.boxplot(column='ctr_percent', by='platform', ax=axes[0])
axes[0].set_title('CTR by Platform', fontweight='bold')
axes[0].tick_params(axis='x', rotation=30)
campaign_df.boxplot(column='engagement_score', by='objective', ax=axes[1])
axes[1].set_title('Engagement by Objective', fontweight='bold')
axes[1].tick_params(axis='x', rotation=30)
plt.suptitle('')
plt.tight_layout()
plt.savefig('eda_viz5_boxplots.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# VIZ 6: Interactive Plotly — ROI by Region
fig = px.box(campaign_df, x='region', y='roi_percent', color='platform',
             title='ROI by Region & Platform', template='plotly_dark')
fig.show()
```

```python
# VIZ 7: Startups — Brand Persona & Stage
stages = ['Idea','Seed','Series A','Series B','Growth','Mature']
personas = ['Minimalist','Vibrant','Luxury','Bold','Friendly','Professional']
startups_df = pd.DataFrame({
    'industry': np.random.choice(['Tech','Fashion','Food','Finance','Health','Education'], 300),
    'brand_persona': np.random.choice(personas, 300),
    'stage': np.random.choice(stages, 300),
    'founding_year': np.random.randint(2010, 2025, 300),
    'employees': np.random.randint(1, 500, 300)
})
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
colors = ['#2E86AB','#A23B72','#F18F01','#C73E1D','#3B1F2B','#44BBA4']
stage_counts = startups_df['stage'].value_counts()
axes[0].barh(stage_counts.index, stage_counts.values, color=colors[:len(stage_counts)])
axes[0].set_title('Startups by Funding Stage', fontweight='bold')
persona_counts = startups_df['brand_persona'].value_counts()
axes[1].bar(persona_counts.index, persona_counts.values, color=colors[:len(persona_counts)])
axes[1].set_title('Brand Persona Distribution', fontweight='bold')
axes[1].tick_params(axis='x', rotation=30)
plt.tight_layout()
plt.savefig('eda_viz7_startups.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# VIZ 8: Class Balance Check
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
platform_counts = campaign_df['platform'].value_counts()
axes[0].bar(platform_counts.index, platform_counts.values, color=colors)
axes[0].set_title('Campaign Count by Platform', fontweight='bold')
axes[0].tick_params(axis='x', rotation=30)
balance = platform_counts.max() / platform_counts.min()
axes[0].text(0.02, 0.95, f'Balance Ratio: {balance:.2f}x', transform=axes[0].transAxes,
             bbox=dict(facecolor='wheat', alpha=0.5))
obj_counts = campaign_df['objective'].value_counts()
axes[1].bar(obj_counts.index, obj_counts.values, color=colors)
axes[1].set_title('Campaign Count by Objective', fontweight='bold')
axes[1].tick_params(axis='x', rotation=30)
plt.tight_layout()
plt.savefig('eda_viz8_class_balance.png', dpi=150, bbox_inches='tight')
plt.show()
```
